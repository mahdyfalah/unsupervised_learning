```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
data1 = np.genfromtxt('dataset1_nocluster15.csv', delimiter=',')
data1_labels = data1[1:, 2:]
data1 = data1[1:, :2]

data2 = np.genfromtxt('dataset2_nocluster2.csv', delimiter=',')
data2_labels = data2[1:, 2:]
data2 = data2[1:, :2]

data3 = np.genfromtxt('dataset3_nocluster2.csv', delimiter=',')
data3_labels = data3[1:, 2:]
data3 = data3[1:, :2]

data4 = np.genfromtxt('dataset4_nocluster2.csv', delimiter=',')
data4_labels = data4[1:, 2:]
data4 = data4[1:, :2]
```

```python
#DBSCAN on Data 1
data1 = StandardScaler().fit_transform(data1)
db = DBSCAN(eps=0.3).fit(data1)

# Plot result
plt.scatter(data1[:, 0], data1[:, 1], c=db.labels_)
```

```python
#DBSCAN on Data 2
data2 = StandardScaler().fit_transform(data2)
db = DBSCAN(eps=0.3).fit(data2)

# Plot result
plt.scatter(data2[:, 0], data2[:, 1], c=db.labels_)
```

```python
#DBSCAN on Data 3
data3 = StandardScaler().fit_transform(data3)
db = DBSCAN(eps=0.3).fit(data3)

# Plot result
plt.scatter(data3[:, 0], data3[:, 1], c=db.labels_)
```

```python
#DBSCAN on Data 4
data4 = StandardScaler().fit_transform(data4)
db = DBSCAN(eps=0.3).fit(data4)

# Plot result
plt.scatter(data4[:, 0], data4[:, 1], c=db.labels_)
```

```python
#KMeans on Data 1
data1 = StandardScaler().fit_transform(data1)
km = KMeans().fit_predict(data1)


# Plot result
plt.scatter(data1[:, 0], data1[:, 1], c=km)
```

```python
#KMeans on Data 2
data2 = StandardScaler().fit_transform(data2)
km = KMeans().fit_predict(data2)


# Plot result
plt.scatter(data2[:, 0], data2[:, 1], c=km)
```

```python
#KMeans on Data 3
data3 = StandardScaler().fit_transform(data3)
km = KMeans().fit_predict(data3)


# Plot result
plt.scatter(data3[:, 0], data3[:, 1], c=km)
```

```python
#KMeans on Data 4
data4 = StandardScaler().fit_transform(data4)
km = KMeans().fit_predict(data4)


# Plot result
plt.scatter(data4[:, 0], data4[:, 1], c=km)
```

```python
#Average link Data 1
data1 = StandardScaler().fit_transform(data1)
average = AgglomerativeClustering(n_clusters=8, linkage='average').fit(data1)


# Plot result
plt.scatter(data1[:, 0], data1[:, 1], c=average.labels_)
```

```python
#Average link Data 2
data2 = StandardScaler().fit_transform(data2)
average = AgglomerativeClustering(n_clusters=2, linkage='average').fit(data2)


# Plot result
plt.scatter(data2[:, 0], data2[:, 1], c=average.labels_)
```

```python
#Average link Data 3
data3 = StandardScaler().fit_transform(data3)
average = AgglomerativeClustering(n_clusters=2, linkage='average').fit(data3)


# Plot result
plt.scatter(data3[:, 0], data3[:, 1], c=average.labels_)
```

```python
#Average link Data 4
data4 = StandardScaler().fit_transform(data4)
average = AgglomerativeClustering(n_clusters=2, linkage='average').fit(data4)


# Plot result
plt.scatter(data4[:, 0], data4[:, 1], c=average.labels_)
```

```python
#EM Data 1
data1 = StandardScaler().fit_transform(data1)
EM = GaussianMixture(n_components=8).fit(data1)


# Plot result
plt.scatter(data1[:, 0], data1[:, 1], c=EM.predict(data1))
```

```python
#EM Data 2
data2 = StandardScaler().fit_transform(data2)
EM = GaussianMixture(n_components=2).fit(data2)


# Plot result
plt.scatter(data2[:, 0], data2[:, 1], c=EM.predict(data2))
```

```python
#EM Data 3
data3 = StandardScaler().fit_transform(data3)
EM = GaussianMixture(n_components=2).fit(data1)


# Plot result
plt.scatter(data3[:, 0], data3[:, 1], c=EM.predict(data3))
```
